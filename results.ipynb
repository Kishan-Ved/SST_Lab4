### Calculating the cyclomatic complexity values for the old and new file

```python
import pandas as pd
from pydriller import Repository

def get_mcc(repo_path, commit_sha, file_path):
        for commit in Repository(repo_path, only_commits=commit_sha).traverse_commits():
            for mod in commit.modified_files:
                old_comp = 0
                if mod.new_path == file_path:
                    for it in mod.methods_before:
                        old_comp += it.complexity
                    print(old_comp)
                    return mod.complexity, old_comp
        return None, None

repo_path = "./faceswap"

for index, row in df.iterrows():
    a, b = get_mcc(repo_path, [row['Commit SHA']], row['New File Path'])
    print(a, b)

    if a is not None and b is not None:
        df.at[index, 'new_file_mcc'] = a
        df.at[index, 'old_file_mcc'] = b

```

### Plots for cyclomatic complexity

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('final.csv')

df['new_file_mcc'] = pd.to_numeric(df['new_file_mcc'], errors='coerce')
df['new_file_mcc'] = df['new_file_mcc'].fillna(0)

df['group'] = (df['Commit SHA'] != df['Commit SHA'].shift()).cumsum()
mcc_sums = df.groupby('group')['new_file_mcc'].sum()

plt.figure(figsize=(12, 5))
plt.plot(mcc_sums.index, mcc_sums.values, linestyle='-')
plt.xlabel('Grouped Commits')
plt.ylabel('Sum of new_file_mcc')
plt.title('Sum of new_file_mcc for Consecutive Commit SHAs')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df[['new_file_mcc', 'old_file_mcc']] = df[['new_file_mcc', 'old_file_mcc']].apply(pd.to_numeric, errors='coerce')

df['mcc_diff'] = (df['new_file_mcc'].fillna(0) - df['old_file_mcc'].fillna(0))

df['group'] = (df['Commit SHA'] != df['Commit SHA'].shift()).cumsum()
mcc_diff_sums = df.groupby('group')['mcc_diff'].sum()

plt.figure(figsize=(12, 5))
plt.plot(mcc_diff_sums.index, mcc_diff_sums.values, linestyle='-')
plt.xlabel('Grouped Commits')
plt.ylabel('Sum of (new_file_mcc - old_file_mcc)')
plt.title('Sum of MCC Differences for Consecutive Commit SHAs')
plt.show()
```
